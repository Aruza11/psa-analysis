```python
import numpy as np
import pandas as pd
import stumps
```

```python
## load data
data = pd.read_csv("~/Documents/Duke/Cynthia Research/KY-analysis-mytrials/KY Recidivism/KY data/kentucky_stumps.csv")
X, Y = data.loc[:,:'current_violence>=1'], data.loc[:,'recid_two_year':]
Y[Y == -1] = 0
cols = data.columns[:-14]
```

```python
Y_recid6 = Y['recid_six_month'].values
Y_recid_drug6 = Y['recid_drug_six_month'].values
Y_recid_violence6 = Y['recid_violence_six_month'].values
Y_recid_F6 = Y['recid_F_six_month'].values
Y_recid_M6 = Y['recid_M_six_month'].values
Y_recid_property6 = Y['recid_property_six_month'].values
```

```python
alpha = [0.001]
recid6 = stumps.stump_features(X, Y_recid6, cols, alpha, 816)
recid_violence6 = stumps.stump_features(X, Y_recid_violence6, cols, alpha, 816)
recid_drug6 = stumps.stump_features(X, Y_recid_drug6, cols, alpha, 816)
recid_property6 = stumps.stump_features(X, Y_recid_property6, cols, alpha, 816)
recid_F6 = stumps.stump_features(X, Y_recid_F6, cols, alpha, 816)
recid_M6 = stumps.stump_features(X, Y_recid_M6, cols, alpha, 816)
```

```python
results = [["Recid Six Month", round(np.mean(recid6['test_auc']),3), round(np.mean(recid6['test_auc']),3) - round(np.mean(recid6['test_auc']),3)],
           ["Violence Six Month", round(np.mean(recid_violence6['test_auc']), 3), round(np.mean(recid_violence6['test_auc']),3) - round(np.mean(recid_violence6['test_auc']),3)],
           ["Drug Six Month", round(np.mean(recid_drug6['test_auc']),3), round(np.mean(recid_drug6['test_auc']),3) - round(np.mean(recid_drug6['test_auc']),3)],
           ["Property Six Month", round(np.mean(recid_property6['test_auc']),3), round(np.mean(recid_property6['test_auc']),3) - round(np.mean(recid_property6['test_auc']),3)],
           ["Felony Six Month", round(np.mean(recid_F6['test_auc']),3), round(np.mean(recid_F6['test_auc']),3) - round(np.mean(recid_F6['test_auc']),3)],
           ["Misdemeanor Six Month", round(np.mean(recid_M6['test_auc']),3), round(np.mean(recid_M6['test_auc']),3) - round(np.mean(recid_M6['test_auc']),3)]]
```

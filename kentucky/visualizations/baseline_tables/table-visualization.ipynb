```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
KY_baseline_data = pd.read_csv('KY Baseline Summary.csv')
KY_interpretable_data = pd.read_csv('KY Interpretable Summary.csv')
FL_baseline_data = pd.read_csv('FL Baseline Summary.csv')
FL_interpretable_data = pd.read_csv('FL Interpretable Summary.csv')
```

## KY Baseline

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Problem', 
                y='Performance', 
                data=KY_baseline_data, 
                hue='Model', 
                style='Time', 
                s=100)
plt.title('Kentucky Baseline Prediction Results', fontsize=15)
plt.xlabel("Recidivism Labels", fontsize=14)
plt.ylabel("Performances - AUC", fontsize=14)
plt.legend(loc='right', 
           bbox_to_anchor=(1.3,0.5),
           fancybox=True, 
           shadow=True)
pass
```

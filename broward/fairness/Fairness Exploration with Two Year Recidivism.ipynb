```python
import numpy as np 
import pandas as pd 
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split


from aequitas.group import Group
from aequitas.bias import Bias
from aequitas.fairness import Fairness
from aequitas.plotting import Plot

%matplotlib inline
```

### Exploration with Aequitas & Logistic Regression 


Read in data 

```python
### load data
data = pd.read_csv("../../../data/broward_data.csv")
data = data.drop(['person_id', 'screening_date'], axis=1)
X = data.loc[:,:'five_year'].values
Y = data['recid_two_year'].values
```

```python
data = pd.read_csv("../../../data/broward_data.csv")
X = data.loc[:,:'five_year']
Y = data['recid_two_year']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.3, shuffle = True, random_state = 5829)

print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
clf = LogisticRegression(class_weight = 'balanced', solver='liblinear', random_state=0)
clf.fit(X_train.drop(['person_id', 'screening_date'], axis=1),
        Y_train)
preds = clf.predict(X_test.drop(['person_id', 'screening_date'], axis=1))
X_test.loc[:,"score"] = preds
X_test.loc[:,"label_value"] = Y_test
# acc = clf.score(X_test,Y_test)
X_test
```

```python
test_fairness = X_test[["person_id", "screening_date", "sex", "score", "label_value"]]
test_fairness = test_fairness.rename({"person_id": "entity_id"}, axis="columns")

# map numeric columns to string names
sex_decoder = {0: "male",
               1: "female"}

test_fairness = test_fairness.replace({"sex": sex_decoder})
```

```python
g = Group()
xtab, _ = g.get_crosstabs(test_fairness, attr_cols=['sex'])
```

Calculated counts across sample population groups

```python
absolute_metrics = g.list_absolute_metrics(xtab)
xtab[[col for col in xtab.columns if col not in absolute_metrics]]
```

Calculated absolute metrics across sample population groups 

tpr = true positive rate 

tnr = true negative rate 

for = false omission rate: incorrectly excluded from intervention group 

fdr = false discovery rate: incorrectly included in intervention group 

fpr = false positive rate 

fnr = false negative rate 

npv =  

ppr = predicted positive rate 

pprev = predicted prevalence rate 

prev = prevalence?

```python
xtab[['attribute_name', 'attribute_value'] + absolute_metrics].round(2)
```

```python
aqp = Plot()
```

```python
fnr = aqp.plot_group_metric(xtab, 'fnr', min_group_size = .05)
fpr = aqp.plot_group_metric(xtab, 'fpr', min_group_size = .05)
```

```python
p = aqp.plot_group_metric_all(xtab, 
#                               metrics=['ppr','pprev','fnr','fpr'], 
                              ncols=4)
```

# Bias


```python
b = Bias() 
```

```python
bdf = b.get_disparity_predefined_groups(xtab, 
                                        original_df=test_fairness, 
                                        ref_groups_dict={'sex':'male'}, 
                                        alpha=0.05, 
#                                         check_significance=True, 
#                                         mask_significance=True
                                        )
bdf.style
```

```python
# View disparity metrics added to dataframe
bdf[['attribute_name', 'attribute_value'] +
     b.list_disparities(bdf) + b.list_significance(bdf)].style
```

```python
fpr_disparity = aqp.plot_disparity(bdf, group_metric='fpr_disparity', attribute_name='sex', significance_alpha=0.05)
```

```python
prec_disparity = aqp.plot_disparity_all(bdf, metrics=['precision_disparity'], significance_alpha=0.05)
```

```python
attr_disp = aqp.plot_disparity_all(bdf, attributes=['sex'], significance_alpha=0.05)
```

# Fairness



Also assessed:
Type I Parity: Fairness in both FDR Parity and FPR Parity

Type II Parity: Fairness in both FOR Parity and FNR Parity

Equalized Odds: Fairness in both FPR Parity and TPR Parity

Unsupervised Fairness: Fairness in both Statistical Parity and Impact Parity

Supervised Fairness: Fairness in both Type I and Type II Parity

Overall Fairness: Fairness across all parities for all attributes

```python
f = Fairness()
fdf = f.get_group_value_fairness(bdf)
```

```python
parity_determinations = f.list_parities(fdf)
fdf[['attribute_name', 'attribute_value'] + absolute_metrics + b.list_disparities(fdf) + parity_determinations].style
```
